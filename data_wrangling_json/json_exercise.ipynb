## JSON exercise

Using data in file 'data/world_bank_projects.json',
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
import pandas as pd
import json
from pandas.io.json import json_normalize
import numpy as np
```

```python
# read and load JSON file into dataframe, then inspect dataframe structure

filename = 'data/world_bank_projects.json'
df = pd.read_json(filename)
df.info()
```

# 1. Find the 10 countries with most projects

Since each project is paired with a country name, we can count by the occurence of each country name to find the top 10 countries with most projects.

```python
# Count occurence of each country name

countryname_counts = df.countryname.value_counts()
print(countryname_counts.head(10))
```

# 2. Find the top 10 major project themes (using column 'mjtheme_namecode')

The column 'mjtheme_namecode' is a dict of code and name of projects. Because some row of the name are empty, we need to count the Occurence of project code instead of name. 

```python
# load json as string

df = json.load((open('data/world_bank_projects.json')))
```

```python
# Normalize the JSON string to table from nested element. 
# After inspection, it shows some row of name are empty.

df = json_normalize(df, 'mjtheme_namecode')
df.head(10)
```

```python
# Save the total count of each project code:
project_count = df.code.value_counts()

# Select the top ten project theme and inspect:
top_ten_major_project_theme = project_count.head(10)
top_ten_major_project_theme
```

# 3. Create a dataframe with the missing names filled in.

By sorting projects' code and name, all rows with empty project name will be on top of those with name filled. First we filled the empty field with NaN, then we can apply the .fillna() method with 'bfill' Backward fill to fill all NaN with appropriate project names.

```python
# Sort the dataframe by code and name, then inspect dataframe

df = df.sort_values(by=['code','name'])
df.head(10)
```

```python
# Fill empty field with NaN
df.name[df.name == ''] = np.nan

# Fill NaN with backward fill method
df_filled = df.fillna(method='bfill')
df_filled.head(50)
```
